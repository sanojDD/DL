```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1)  # No activation because it's regression
])
```

```python
def build_model(hp):
    model = Sequential()
    units = hp.Int('units', min_value=8, max_value=128, step=8)
    model.add(Dense(units=units, activation='relu', input_dim=X_train.shape[1]))
    model.add(Dense(1))  # No activation for regression output
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
print(X_train.shape, y_train_scaled.shape)
print(X_test.shape, y_test_scaled.shape)
```

```python
print(X_test_scaled.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit scaler only on the training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform test data with the same scaler
X_test_scaled = scaler.transform(X_test)
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective="val_mae",
    max_trials=10,
    executions_per_trial=1,
    directory="keras_tuner_house",
    project_name="house_price"
)
```

```python
print(X_test_scaled.shape)
```

```python
tuner.search(
    X_train_scaled, y_train,
    validation_data=(X_test_scaled, y_test),
    epochs=50,
    batch_size=32
)
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model = tuner.get_best_models(num_models=1)[0]
```

```python
history2 = model.fit(X_train_scaled, y_train_scaled, batch_size=32, epochs=100, initial_epoch=6 ,validation_data=(X_test, y_test))
```
